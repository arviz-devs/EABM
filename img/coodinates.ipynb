```python
import matplotlib.pyplot as plt
import numpy as np
import arviz_plots as azp
```

```python
azp.style.use("arviz-variat")
```

```python
# Enter x and y coordinates of points and colours
xs = [2, -3]
ys = [3, 1]
colours = ['C0', 'C1']

# Select length of axes and the space between tick labels
xmin, xmax, ymin, ymax = -4, 4, -4, 4

# Plot points
fig, ax = plt.subplots(figsize=(8, 8))
ax.scatter(xs, ys, c=colours)

# Draw lines connecting points to axes
for x, y, c in zip(xs, ys, colours):
    ax.plot([x, x], [0, y], c=c, ls='--', lw=1.5, alpha=0.5)
    ax.plot([0, x], [y, y], c=c, ls='--', lw=1.5, alpha=0.5)

# Set identical scales for both axes
ax.set(xlim=(xmin-1, xmax+1), ylim=(ymin-1, ymax+1), aspect='equal')

# # Set bottom and left spines as x and y axes of coordinate system
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')

# Create 'x' and 'y' labels placed at the end of the axes
ax.set_xlabel('x',  labelpad=-24, x=1.03)
ax.set_ylabel('y',  labelpad=-21, y=1.02, rotation=0)

# Create custom major ticks to determine position of tick labels
x_ticks = np.arange(xmin, xmax+1)
y_ticks = np.arange(ymin, ymax+1)
ax.set_xticks(x_ticks)
ax.set_yticks(y_ticks[y_ticks != 0])

# Create minor ticks placed at each integer to enable drawing of minor grid
# lines: note that this has no effect in this example with ticks_frequency=1
ax.set_xticks(np.arange(xmin, xmax+1), minor=True)
ax.set_yticks(np.arange(ymin, ymax+1), minor=True)

# Draw major and minor grid lines
ax.grid(which='both', color='grey', linewidth=1, linestyle='-', alpha=0.2)

# Draw arrows
ax.plot((1), (0), marker='>', transform=ax.get_yaxis_transform(), markersize=7, color='k', clip_on=False, zorder=3)
ax.plot((0), (1), marker='^', transform=ax.get_xaxis_transform(), markersize=7, color='k', clip_on=False, zorder=3)

ax.text(2.15, 3.15, '(2, 3)', color=colours[0])
ax.text(-3.15, 1.15, '(-3, 1)', color=colours[1])

plt.savefig('cartesian_coordinates.png', dpi=300)
```

```python
# Plot points
fig, ax = plt.subplots(figsize=(8, 4))
ax.scatter(xs, ys, c=colours)


# Create 'x' and 'y' labels placed at the end of the axes
ax.set_xlabel('x')
ax.set_ylabel('y')

# Draw major and minor grid lines
ax.grid(which='both', color='grey', linewidth=1, linestyle='-', alpha=0.2)

ax.text(2.15, 3.15, '(2, 3)', color=colours[0])
ax.text(-3.15, 1.15, '(-3, 1)', color=colours[1])

plt.savefig('cartesian_coordinates_shifted.png', dpi=300)
```

```python


# Create a polar coordinate plot
_, axes = plt.subplots(figsize=(6, 6), subplot_kw={'projection': 'polar'}) 

# Create data for polar coordinates
r = np.array([0.6, 1.4])
theta = [0.2, 2]


axes.plot(theta[0], r[0], 'C0o')
axes.plot(theta[1], r[1], 'C1o')
axes.set_rticks(np.arange(0.2, 1.6, 0.4))
axes.set_xlabel('x axis (circular)')
axes.set_ylabel('y axis (radial)', labelpad=35)

axes.set_title('Polar Coordinate System')
plt.savefig('polar_coordinates.png', dpi=300)
```
